```python
from rank_bm25 import BM25Okapi, BM25Plus
from sudachipy import Dictionary, SplitMode # トークン化に利用
from pprint import pprint
import numpy as np
tokenizer = Dictionary().create()
def extract_nouns(text): # 名詞を抽出
    tokens = tokenizer.tokenize(text, SplitMode.C)
    return [m.surface() for m in tokens if m.part_of_speech()[0]=='名詞']

corpus = [
    '順に秤量し、ビーカーに加える。各秤量値は実験ノートに記録しておく。',
    'スターラーで軽く攪拌して溶解・混合を促進する。',
    '順番に加え、それぞれ秤量値を記録する。',
    '同じビーカー内で再度スターラー攪拌を行う。',
    'ビーカーを超音波分散機にセットし分散処理を行う。'
]
tokenized_texts =[]
for text in corpus:
    tokenized_texts.append(extract_nouns(text))
pprint(tokenized_texts)
# Index作成
bm25index = BM25Plus(tokenized_texts) # index作成

# 検索
tokenized_query = extract_nouns('攪拌混合')
print('Query: ', tokenized_query)
doc_scores = bm25index.get_scores(tokenized_query)
print(doc_scores) # 各text(入力順)に対するスコア
print(corpus[np.argmax(doc_scores)]) # 1位text
```
